```python
import random

class Card(object):
    def __init__(self, suit, val):
        self.suit = suit
        self.value = val

    # Implementing build in methods so that you can print a card object
    def __unicode__(self):
        return self.show()
    def __str__(self):
        return self.show()
    def __repr__(self):
        return self.show()
    def show(self):
        val = self.value
        return "{} of {}".format(val, self.suit)
    
class Deck(object):
    def __init__(self):
        self.cards = []
        self.build()

    # Generate 52 cards
    def build(self):
        self.cards = []
        for suit in ["\u2660","\u2665","\u2666","\u2663"]:
            for val in range(1,14):
                self.cards.append(Card(suit, val))

    # Return the top card
    def deal(self):
        random.shuffle(self.cards)
        return self.cards.pop()

class Player(object):
    def __init__(self, name):
        self.name = name
        self.hand = []

    # Draw n number of cards from a deck
    # Returns true in n cards are drawn, false if less then that
    def draw(self, deck, num=2):
        for _ in range(num):
            card = deck.deal()
            if card:
                self.hand.append(card)
            else: 
                return False
        return True

    # Display all the cards in the players hand
    def showHand(self):
        C1 = self.hand[0]
        C_total = C1.value
        if C1.value ==1 or C1.value >10 :
            if C1.value == 1:
                C1.value == 'A'
            elif C1.value == 11:
                C1.value = 'J'
            elif C1.value == 12:
                C1.value = 'Q'
            elif C1.value == 13:
                C1.value = 'K'
            print("""\
                        ┌─────────┐
                        │{}        │
                        │         │
                        │         │
                        │    {}    │
                        │         │
                        │         │
                        │        {}│
                        └─────────┘
                        """.format(C1.value, C1.suit, C1.value ) )
        elif C1.value <10 and C1.value>1 :
            print("""\
                        ┌─────────┐
                        │{}        │
                        │         │
                        │         │
                        │    {}    │
                        │         │
                        │         │
                        │       {} │
                        └─────────┘
                        """.format(C1.value, C1.suit, C1.value))
        C2 = self.hand[1]
        C_total = C_total + C2.value
        if C2.value ==1 or C2.value > 10 :
            if C2.value == 1:
                C2.value = 'A'
            elif C2.value == 11:
                C2.value = 'J'
            elif C2.value == 12:
                C2.value = 'Q'
            elif C2.value == 13:
                C2.value = 'K'
            print("""\
                        ┌─────────┐
                        │{}        │
                        │         │
                        │         │
                        │    {}    │
                        │         │
                        │         │
                        │        {}│
                        └─────────┘
                        """.format(C2.value, C2.suit, C2.value) )
        elif C2.value <10 and C2.value>1 :
            print("""\
                        ┌─────────┐
                        │{}        │
                        │         │
                        │         │
                        │    {}    │
                        │         │
                        │         │
                        │       {} │
                        └─────────┘
                        """.format(C2.value, C2.suit, C2.value))
        return C_total

    def discard(self):
        return self.hand.pop()
count = [0,0]
while True:
    total = []
    myDeck = Deck()
    # deck.show()
    for i in range(0,2):
        print("Player "+str(i+1)+" turn: ")
        player_1 = Player("player_1")
        player_1.draw(myDeck, 2)
        P1 = player_1.showHand()
        print("Player "+str(i+1)+" total: "+str(P1))
        total.append(P1)
    winningID = total[0]
    if winningID < total[1]:
        winningID = total[1]
        print("Player 2 is Winner : "+str(winningID))
        count[1] += 1
    else:
        print("Player 1 is Winner : "+str(winningID))
        count[0] += 1
    print("Overall Score is Player 1: "+str(count[0])+" and PLayer 2: "+str(count[1]))
    repeat = input("Do you want to repeat the game :")
    if repeat == 'Y':
        continue
    elif repeat == 'N':
        break
print("ok")    
```
